Был дополнительно построен график по росту дохода компании.
Видно, что у компания Shmuber намного хуэе доход, по сравнению с другими компаниями.
Дополнительно я бы запросил информацию по росту клиентов в данной компании.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C://Users/Tem/Downloads/test.csv')
legends = list()
for name, group in df.groupby('company'):
    legends.append(name)
    plt.plot(group['year'], group['revenue, $K'])
plt.legend(legends)
```
