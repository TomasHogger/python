```python
import pandas as pd

#Считываем данные из файлов
student_df = pd.read_csv('students.csv')
course_df = pd.read_csv('courses.csv').drop(columns=['Unnamed: 0'])
progress_phases_df = pd.read_csv('progress_phases.csv')
progresses_df = pd.read_csv('progresses.csv')
course_contents_df = pd.read_csv('course_contents.csv')
```

```python
#Формируем три DataFrame таблицы

#Курсы и подробности о них
courses_full_df = course_contents_df.join(course_df.set_index('id'), on='course_id', how='left')
#Прогресс и подробности о нём
progress_full_df = progress_phases_df.join(progresses_df.set_index('id'), on='progress_id', how='left')
#Полный DataFrame
main_df = pd.merge(courses_full_df, progress_full_df, on=['course_id', 'lesson_number',  'module_number'])
```

```python
from datetime import datetime
from datetime import date

#Функция нахождения возраста
def calculate_age(born):
    today = date.today() 
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

#Формируем DataFrame из студентов, у которых указан возраст
student_with_age_df = student_df[student_df['birthday'].notna()]
#Приводим дату рождения рождения к классу Datetime
student_with_age_df['birthday'] = [datetime.strptime(x, "%Y-%m-%d").date() for x in student_with_age_df['birthday']]
#Рассчитываем возраст и записываем в колонку age
student_with_age_df['age'] = student_with_age_df['birthday'].apply(calculate_age)
```

```python
print('Минимальный, максимальный, средний, медианный возраст студентов')
student_with_age_df['age'].aggregate(['min', 'max', 'mean', 'median'])
```

Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
#Создаём колонку start_date_ym в main_df и кладём туда значения из колонки start_date в формате год-месяц
main_df['start_date_ym'] = ['-'.join(x.split('-')[0:2]) for x in main_df['start_date']]
#Выбираем только те записи, где есть ДЗ
df = main_df[(main_df['is_homework'] == True)]

#Словарь, где ключ - это название курса, а значение - количество студентов на этом курсе, которые сдали первое ДЗ, по месяцам
course_student_count_dict = dict()

#Группируем по названию курса
for name, group in df.groupby('title'):
    #Сортируем по названию модуля и номеру урока, чтобы выбрать самый первый модуль и урок, где есть ДЗ
    group = group.sort_values(by=['module_number', 'lesson_number'])
    #Выбираем только те записи, которые соответствуют самой первой по столбцам module_number и lesson_number
    group = group[(group['module_number'] == group['module_number'].values[0]) & (group['lesson_number'] == group['lesson_number'].values[0])]
    #Группируем по дате сдачи ДЗ и считаем количество студентов по датам
    count_student_by_month_df = group.groupby('start_date_ym')['student_id'].count().to_frame()
    #Записываем в словарь 
    course_student_count_dict[name] = count_student_by_month_df
    print(name, count_student_by_month_df, '\n\n')
```

Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы и временем окончания выполнения последней домашней работы в модуле

```python
#Групируем данные по названию курса, названию модуля и id студента
title_md_si_group = main_df.groupby(['title', 'module_title', 'student_id'])

#Словарь, где ключ - это курс и модуль, а значение - это массив времени в секундах, затраченное на прохождение модуля
course_times_dict = dict()

for name, group in title_md_si_group:
    #Если в данном модуле нет ДЗ, то пропускаем
    group_homework = group[group['is_homework'] == True]
    if len(group_homework['start_date'].values) == 0:
        continue

    #Сортируем по дате сдачи ДЗ студентом и выбираем колонку start_date 
    sort_start_date = group_homework.sort_values('start_date')['start_date']
    #Сортируем по дате принятия ДЗ в обратном порядке у студента и выбираем колонку finish_date 
    sort_finish_date = group_homework.sort_values('finish_date', ascending=False)['finish_date']
    
    #Паттерн написания даты
    date_format = '%Y-%m-%d %H:%M:%S'
    
    #Выбираем дату последнего принятия ДЗ
    finish_date = sort_finish_date.values[0]
    
    #Если нет даты принятия ДЗ, то не учитываем данного студента 
    if finish_date is np.nan:
        continue
        
    
    #Берём самую первую дату сдачи ДЗ студентом в этом модуле и убираем из неё TimeZone
    start_date = sort_start_date.values[0].split('+')[0].split('.')[0]
    
    #Пробуем спарсить дату первой дамашки по первому паттерну даты, если не получится, то парсим по второму (в тип datetime)
    start_date = datetime.strptime(start_date, date_format)
        
    #Убираем TimeZone и даты последнего принятия ДЗ
    finish_date = finish_date.split('+')[0].split('.')[0]
    
    #Пробуем спарсить дату последнего принятия ДЗ по первому паттерну даты,
    #если не получится, то парсим по второму (в тип datetime)
    finish_date = datetime.strptime(finish_date, date_format)

    #Берём только название курса и название модуля
    name = name[0:2]
        
    #Берём из словаря course_times_dict массив под ключом name (либо пустой массив)
    times = course_times_dict.get(name, list())
    #Добавляем в массив разницу между finish_date и start_date в секундах
    time_diff = finish_date - start_date
    times.append(int(time_diff.total_seconds()))
    #Добавляйем обратно в словарь
    course_times_dict[name] = times
```

Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам (12 месяцев, январь-декабрь) для каждого курса.

```python
#словарь, где ключ - это курс, а значение - это словарь, где ключ - это месяц,
#а значение - это список из времени, затраченного на выполнение ДЗ, в секундах 
course_month_times_dict = dict()

#берём только дз и группируем по курсу
for name, group in main_df[(main_df['is_homework'] == True)].groupby('title'):
    #словарь, где ключ - это месяц, а значение - это список
    #из времени, затраченного на выполнение ДЗ, в секундах 
    month_times_dict = dict()
    
    #группируем по модулю, номеру ДЗ и студенту
    for name2, group2 in group.groupby(['module_number', 'lesson_number', 'student_id']):
        
        #берём дату принятия ДЗ
        group2 = group2.sort_values('finish_date', ascending=False)
        finish_date = group2['finish_date'].values[0]
    
        #если даты нет, то не учитываем эту запись
        if finish_date is np.nan:
            continue
        
        #берём дату первой сдачи ДЗ (либо дату начала ДЗ)
        group2 = group2.sort_values('start_date')
        start_date = group2['start_date'].values[0].split('+')[0].split('.')[0]
        #делим дату на составляющие и приводим к числам (год, месяц, день)
        sy, sm, sd = [int(i) for i in start_date.split(' ')[0].split('-')]

        #убираем лишнее из даты принятия ДЗ
        finish_date = finish_date.split('+')[0].split('.')[0]
        finish_date_ymd = finish_date.split(' ')[0]
        
        #делим дату на составляющие и приводим к числам (год, месяц, день)
        fy, fm, fd = [int(i) for i in finish_date_ymd.split('-')]
        
        #промежуточная дата на месяц больше даты сдачи ДЗ первый раз (необходимо, если сдача ДЗ растянулось на период > месяца)
        iy, im, id = sy, 1, 1
        if sm == 12:
            iy += 1
        else:
            im = sm + 1
    
        #формат даты
        date_format = '%Y-%m-%d %H:%M:%S'
    
        #цикл, пока промежуточная дата не обгонит дату сдачи ДЗ
        #данный цикл необходимо, чтобы расчитывать время затраченно на ДЗ по месяцам. 
        #пример: ученик начал 03-01-2019, а закончил 05-02-2019.
        #мы 27 дней занесём в январь, а 5 дней в февраль.
        while iy < fy or (iy == fy and im <= fm):
            #берём список со временем для месяца
            times = month_times_dict.get(sm, list())
            #расчитываем разницу между датой сдачи дз (либо началом месяца) и началом следующего месяца
            start_date = datetime.strptime(start_date, date_format)
            inter_date = datetime(iy, im, id)
            times.append(int((inter_date - start_date).total_seconds()))
            month_times_dict[sm] = times
            #устанавливаем в переменную start_date дату, которая была промежуточная,
            #а промежуточную дату увеличиваем на месяц. это нужно,
            #если ДЗ затянулось больше, чем на 2 месяца
            start_date = "%d-%02d-0%d 00:00:00" % (iy, im, id)
            sy, sm, sd = iy, im, id
        
            if im == 12:
                iy += 1
                im = 1
            else:
                im += 1
            
        #расчитываем разницу между первой сдачей ДЗ (либо датой,
        #которая была в последний раз промежуточная) и принятием ДЗ
        times = month_times_dict.get(sm, list())
        start_date = datetime.strptime(start_date, date_format)
        finish_date = datetime.strptime(finish_date, date_format)
        times.append(int((finish_date - start_date).total_seconds()))
        month_times_dict[sm] = times
    course_month_times_dict[name] = month_times_dict
```
