–§—É–Ω–∫—Ü–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (-‚àû, -1) –∏ —É–±—ã–≤–∞–µ—Ç –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (-1, ‚àû) <br><br> 

```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
```

```python
def explore_function(x, f, range=None):
    print(f.subs(x, -x))
    print(solve(f))
    print(solve_univariate_inequality(f>0, x))
    print(solve_univariate_inequality(f<0, x))
    if (range==None):
        plot(f)
    else:
        plot(f, range)
```

```python
x = Symbol('x')
f = -5*x**2-10*x+100
explore_function(x,f)
```

```python
x = Symbol('x')
f = 2*ln(x)+5
explore_function(x,f,(x,0,100))
```

```python
x = Symbol('x')
f = 3*sqrt(x)-15
explore_function(x,f,(x,0,100))
```

```python
x = Symbol('x')
f = 2*exp(0.2*x)-1
explore_function(x,f)
```

```python
x = Symbol('x')
f = 5*x+1.5
explore_function(x,f)
```
